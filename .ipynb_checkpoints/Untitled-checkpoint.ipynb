```python
import pandas as pd

def categorize_columns(df):
    """
    Function to separate categorical and numerical columns in a dataset.

    Args:
        df (pd.DataFrame): Input dataset

    Returns:
        dict: A dictionary with keys 'categorical' and 'numerical' containing column names
    """
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
    
    return {
        'categorical': categorical_cols,
        'numerical': numerical_cols
    }
```

```python
# Example Usage
data = pd.read_csv('preproced_data.csv')
columns = categorize_columns(data)

print("Categorical Columns:", columns['categorical'])
print("Numerical Columns:", columns['numerical'])
```

```python
import pandas as pd

def load_dataset(file_path):
    """
    Load the dataset from a file.

    Parameters:
        file_path (str): Path to the file (CSV or Excel).

    Returns:
        pd.DataFrame: Loaded DataFrame or None if an error occurs.
    """
    try:
        if file_path.endswith('.csv'):
            data = pd.read_csv(file_path,encoding='latin1')
        elif file_path.endswith('.xlsx'):
            data = pd.read_excel(file_path)
        else:
            print("Unsupported file format. Please use .csv or .xlsx.")
            return None
        print(f"Dataset loaded successfully with {data.shape[0]} rows and {data.shape[1]} columns.")
        return data
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}.")
    except pd.errors.EmptyDataError:
        print("Error: The file is empty.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


def highest_roi(data):
    """
    Calculate the top 10 areas or cities with the highest ROI for rental properties.

    Parameters:
        data (pd.DataFrame): Dataset containing 'totalRent' and 'baseRent' columns.

    Returns:
        pd.Series: Top 10 cities by ROI.
    """
    try:
        data['ROI'] = (data['totalRent'] * 12) / data['baseRent'] * 100
        top_roi_areas = data.groupby('city')['ROI'].mean().sort_values(ascending=False).head(10)
        return top_roi_areas
    except KeyError as e:
        print(f"Missing required column: {e}")
    except Exception as e:
        print(f"An error occurred while calculating ROI: {e}")


def common_rental_price_range(data):
    """
    Determine the most common rental price range.

    Parameters:
        data (pd.DataFrame): Dataset containing 'totalRent' column.

    Returns:
        str: Most common rental price range.
    """
    try:
        bins = [0, 500, 1000, 1500, 2000, 3000, 5000]
        data['rent_range'] = pd.cut(data['totalRent'], bins=bins, labels=['0-500', '501-1000', '1001-1500', '1501-2000', '2001-3000', '3001-5000'])
        most_common_range = data['rent_range'].value_counts().idxmax()
        return most_common_range
    except KeyError:
        print("Error: 'totalRent' column is missing in the dataset.")
    except Exception as e:
        print(f"An error occurred while calculating rental price range: {e}")


def top_rent_increases(data):
    """
    Identify the top 10 ZIP codes with the highest rent increases.

    Parameters:
        data (pd.DataFrame): Dataset containing 'geo_plz', 'totalRent', and optionally 'yearConstructed' columns.

    Returns:
        pd.Series: Top 10 ZIP codes by average rent increase.
    """
    try:
        if 'yearConstructed' in data.columns:
            data['year'] = data['yearConstructed']
        rent_increase = data.groupby(['geo_plz', 'year'])['totalRent'].mean().pct_change().dropna()
        top_zip_codes = rent_increase.groupby('geo_plz').mean().sort_values(ascending=False).head(10)
        return top_zip_codes
    except KeyError as e:
        print(f"Missing required column: {e}")
    except Exception as e:
        print(f"An error occurred while calculating rent increases: {e}")


def most_common_heating_type(data):
    """
    Determine the most commonly selected heating type.

    Parameters:
        data (pd.DataFrame): Dataset containing 'heatingType' column.

    Returns:
        str: Most common heating type.
    """
    try:
        common_heating = data['heatingType'].value_counts().idxmax()
        return common_heating
    except KeyError:
        print("Error: 'heatingType' column is missing in the dataset.")
    except Exception as e:
        print(f"An error occurred while determining the most common heating type: {e}")


def avg_rent_per_sqft_top_cities(data):
    """
    Calculate the average rent per square foot in the top 5 cities.

    Parameters:
        data (pd.DataFrame): Dataset containing 'totalRent', 'livingSpace', and 'city' columns.

    Returns:
        pd.Series: Average rent per square foot for the top 5 cities.
    """
    try:
        data['rent_per_sqft'] = data['totalRent'] / data['livingSpace']
        top_cities = data.groupby('city')['totalRent'].count().sort_values(ascending=False).head(5).index
        avg_rent = data[data['city'].isin(top_cities)].groupby('city')['rent_per_sqft'].mean()
        return avg_rent
    except KeyError as e:
        print(f"Missing required column: {e}")
    except Exception as e:
        print(f"An error occurred while calculating rent per square foot: {e}")
```
